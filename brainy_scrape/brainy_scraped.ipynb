```python
from cloudscraper import create_scraper # import to create scraper object
from bs4 import BeautifulSoup # import to parse response
import pandas as pd # import for working with dataframes
```

```python
scraper = create_scraper() # create scraper object
```

```python
response = scraper.get('https://www.brainyquote.com/topics/motivational-quotes') # obtain page 1 to construct scrapping
```

```python
a = soup.select('body main .oncl_a') # obtain all author information
```

```python
q = soup.select('body main .b-qt') # obtain all quote information
```

```python
print(len(a)) # ensure lengths are equal to one another and to website page 1 quote number
print(len(q))
```

```python
author = a[0].text # define variables that access each target element
alink = a[0]['href']
quote = q[0].text.strip()
qlink = q[0]['href']
```

```python
alist = [] # define function to obtain author target items, create dictonary of items, then dict to list
def getauthor(a):
    i = 0
    for item in a:
        aitem = a[i]
        author = aitem.text
        alink = aitem['href']
        aitems = {'author': author,
                'author_link': alink}
        alist.append(aitems)
        i += 1
```

```python
len(alist) # ensure 
```

```python
qlist = []
def getquote(q):
    j = 0
    for itemq in q:
        qitem = q[j]
        quote = qitem.text.strip()
        qlink = qitem['href']
        qitems = {'quote': quote,
                'quote_link': qlink}
        qlist.append(qitems)
        j += 1
```

```python
getauthor(a) # verify defined functions work and output looks well formatted
getquote(q)
print(alist)
print(qlist)
```

```python
alist = [] # pagination through pages 1-5 = all pages scrapped
qlist = []
for n in range(1, 6):
    response = scraper.get(f'https://www.brainyquote.com/topics/motivational-quotes_{n}')
    soup = BeautifulSoup(response.text, 'lxml')
    a = soup.select('body main .oncl_a')
    q = soup.select('body main .b-qt')
    getauthor(a)
    getquote(q) 
    print(n) # print to look at inner workings of loop
```

```python
author_df = pd.DataFrame(alist) # create dataframe from author list
author_df.head()
```

```python
quote_df = pd.DataFrame(qlist) # create dataframe from quote list
quote_df.head()
```

```python
df = quote_df.join(author_df) # merge, how = left by default, into one dataframe
df.tail()
```

```python
brainy_quotes = df.to_excel('brainy_quotes.xlsx', index = False) # export dataframe as excel file
```
