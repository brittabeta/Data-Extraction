```python
from pytrials.client import ClinicalTrials
import pandas as pd
from datetime import date
```

```python
ct = ClinicalTrials()
```

```python
cancer_fields = ct.get_study_fields(
    search_expr="malignant+cancer",
    fields=['NCTId', 'Phase', 'StudyType', 'ConditionMeshTerm', 'PrimaryOutcomeMeasure', 'StdAge', 'InterventionType', 
            'InterventionName', 'InterventionOtherName', 'InterventionMeshTerm', 'InterventionAncestorTerm',
            'LeadSponsorName', 'LeadSponsorClass', 'ResponsiblePartyInvestigatorAffiliation', 'OverallOfficialAffiliation', 
            'CollaboratorName', 'CollaboratorClass', 'BriefSummary'],
    max_studies=1000,
    fmt='csv')
```

```python
df = pd.DataFrame(cancer_fields)
df.head()
```

```python
df.columns = df.loc[0]
df.drop(index=0, inplace= True)
```

```python
datevar = date.today().strftime("%b-%d-%Y")
datevar
```

```python
cancer_trials = df.to_csv(f'cancer_trials_{datevar}.csv', index= False)
```
